```python
L = []

base_url = 'https://www.airlinequality.com/airline-reviews/british-airways/page/'  #to capture other pages
for page in range(0,372,1):   
    print(base_url + str(page))
    r = requests.get(base_url + str(page))
    c = r.content  #r.json()['list']
    soup = BeautifulSoup(c,'html.parser')
    #print(soup.prettify())
    all = soup.find_all('div', {'class':'review-stats'})
  

    for items in all:

            d = {}

            try:
                d['Aircraft']= items.find_all('td',{'class':'review-value'})[-6].text 
            except:
                d['Aircraft'] = None 

            try:
                d['Type_of_Traveller'] = items.find_all('td',{'class':'review-value'})[-5].text
            except:
                d['Type_of_Traveller'] = None


            try:
                d['Seat_type'] = items.find_all('td',{'class':'review-value'})[-4].text
            except:
                d['Seat_type'] = None

            try:
                d['Route'] = items.find_all('td',{'class':'review-value'})[-3].text
            except:
                d['Route'] = None

            try:
                d['Date_flown'] = items.find_all('td',{'class':'review-value'})[-2].text
            except:
                d['Date_flown'] = None

            try:
                d['Recommended'] = items.find_all('td',{'class':'review-value'})[-1].text
            except:
                d['Recommended'] = None


            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header seat_comfort'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Seat_comfort'] = total_star_fill
            except :
                d['Seat_comfort'] = None

            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header cabin_staff_service'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Cabin_service'] = total_star_fill
            except :
                d['Cabin_service'] = None


            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header food_and_beverages'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Food'] = total_star_fill
            except :
                d['food'] = None


            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header inflight_entertainment'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Inflight_entertainment'] = total_star_fill
            except :
                d['Inflight_entertainment'] = None



            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header ground_service'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Ground_service'] = total_star_fill
            except :
                d['Ground_service'] = None


            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header wifi_and_connectivity'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Wifi'] = total_star_fill
            except :
                d['Wifi'] = None

            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header value_for_money'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Value_for_money'] = total_star_fill
            except :
                d['Value_for_money'] = None

            #print(' ')   #To give space after each customer's review

            L.append(d)


        
        
          
    
```
